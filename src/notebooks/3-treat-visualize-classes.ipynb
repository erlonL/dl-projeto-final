```python
# create dataframe from annotations
# for each box 
# index | image_path | label | occluded | x_max | y_max | x_min | y_min

with open('../new_dataset/train/train.yaml', 'r') as file:
    data = yaml.safe_load(file)

rows = []
for i, image in enumerate(data):
    image_path = image['path']
    for box in image['boxes']:
        row = {
            'image_path': image_path,
            'label': box['label'],
            'occluded': box['occluded'],
            'x_max': box['x_max'],
            'y_max': box['y_max'],
            'x_min': box['x_min'],
            'y_min': box['y_min']
        }
        rows.append(row)
df_train = pd.DataFrame(rows)
```

```python
# if starts with 'Red' then 'Red'
# if starts with 'Green' then 'Green'
# if starts with 'Yellow' then 'Yellow'
# examples: GreenStraightLeft, RedLeft, ...
def get_label(label):
    if label.startswith('Red'):
        return 'Red'
    elif label.startswith('Green'):
        return 'Green'
    elif label.startswith('Yellow'):
        return 'Yellow'
    else:
        return 'off'
    
df_train['label'] = df_train['label'].apply(get_label)
df_test['label'] = df_test['label'].apply(get_label)
```

```python
class Annotation:
    def __init__(self, label, occluded, x_max, y_max, x_min, y_min):
        self.label = label
        self.occluded = occluded
        self.x_max = x_max
        self.y_max = y_max
        self.x_min = x_min
        self.y_min = y_min

    def __repr__(self):
        return f"Annotation(label={self.label}, occluded={self.occluded}, x_max={self.x_max}, y_max={self.y_max}, x_min={self.x_min}, y_min={self.y_min})"
```

```python
def plot_sample(base_path, image_path, annotation):
    import matplotlib.pyplot as plt
    import matplotlib.patches as patches
    import matplotlib.image as mpimg

    image = mpimg.imread(f"{base_path}/{image_path}")

    fig, ax = plt.subplots(2)
    ax[0].imshow(image)

    # Create a Rectangle patch
    rect = patches.Rectangle((annotation.x_min, annotation.y_min), 
                           annotation.x_max - annotation.x_min, 
                           annotation.y_max - annotation.y_min, 
                           linewidth=1, edgecolor='r', facecolor='none')

    # Add the patch to the Axes
    ax[0].add_patch(rect)
    
    # Convertendo para inteiros antes de fazer o slice
    y_min = int(annotation.y_min)
    y_max = int(annotation.y_max)
    x_min = int(annotation.x_min)
    x_max = int(annotation.x_max)
    
    # Usando os valores convertidos
    ax[1].imshow(image[y_min:y_max, x_min:x_max])
    ax[1].set_xlim(0, x_max - x_min)
    ax[1].set_ylim(0, y_max - y_min)
    ax[1].set_title("Zoomed In")
    ax[0].set_title("Original Image")
    plt.title(f"Label: {annotation.label}, Occluded: {annotation.occluded}")
    plt.tight_layout()
    plt.show()
```

```python
base_path = "../new_dataset/train/rgb/train"
image_path = df_train.iloc[1220]['image_path']

annotation = Annotation(df_train.iloc[1220]['label'],
                        df_train.iloc[1220]['occluded'],
                        df_train.iloc[1220]['x_max'],
                        df_train.iloc[1220]['y_max'],
                        df_train.iloc[1220]['x_min'],
                        df_train.iloc[1220]['y_min'])
plot_sample(base_path, image_path, annotation)
```
